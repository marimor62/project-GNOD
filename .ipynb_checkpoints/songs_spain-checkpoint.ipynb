# Multiple pages Web scraping

Number-one songs in Spain from 1980 to 1989.

```python
from bs4 import BeautifulSoup
import requests
from random import randint
from time import sleep 
```

```python
# range(start, stop, step) desde 1980 a 1989
iterations = range(0, 10, 1)
```

```python
for i in iterations:
    print(i)
```

```python
#create the full url string based on the page search 
#(and test the urls with the iteration)
url_lst=[f"https://en.wikipedia.org/wiki/List_of_number-one_singles_of_198{i}_(Spain)" for i in iterations]
```

```python
url_lst
```

```python
pages = []

for i in iterations:
    url=f"https://en.wikipedia.org/wiki/List_of_number-one_singles_of_198{i}_(Spain)"
    response = requests.get(url)
    print("Status code: "+ str(response.status_code))
    #to capture the data and fill in the list
    pages.append(response)
    #add a respectful nap 
    wait_time=randint(1,4)
    sleep(wait_time)
```

```python
soup1 = BeautifulSoup(response.content, "html.parser")
soup1
```

```python
#get songs title
soup1.select("tr:nth-child(n) > td:nth-child(2) > a")
```

```python
#get artists
soup1.select("tr:nth-child(n) > td:nth-child(3) > a")
```

```python
titles = []
artists = []
            
for i in range(len(pages)):
    response = requests.get(url_lst[i])
    pages_html = BeautifulSoup(response.text, 'html.parser')
    songs_html=pages_html.select("div.mw-parser-output > table > tbody > tr:nth-child(n) > td:nth-child(2)")
    artists_html=pages_html.select("div.mw-parser-output > table > tbody > tr:nth-child(n) > td:nth-child(3)") 
    for j in songs_html:
        titles.append(j.text)  
    for j in artists_html:
        artists.append(j.text)
```

```python
# they have \n in the artists, so let's remove it
artists = [artist.strip('\n') for artist in artists]
```
